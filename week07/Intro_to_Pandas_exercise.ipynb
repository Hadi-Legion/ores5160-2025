Q3. Data Visualization

	1.	Create a histogram of the pm25 values.\\

	2.	Create a boxplot for the pm25 column.\\


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(airquality_cleaned.columns.tolist())
```

```python
pm25_data = airquality_cleaned[airquality_cleaned['value'] == 'pm25']
plt.figure(figsize=(8,5))
plt.hist(pm25_data['value'], bins=30, edgecolor='red')
plt.title('Distribution of PM2.5 Values')
plt.xlabel('PM2.5')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=pm25_data['value'])
plt.title('Boxplot of PM2.5 Levels')
plt.xlabel('PM2.5')
plt.show()
```
